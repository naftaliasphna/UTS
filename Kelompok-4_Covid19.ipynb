```python
# Melakukan penisahan variabel 
X = data.drop(['DIED'],axis=1)
y = data['DIED']

# Disini dipisahkan kolom "DIED" sebagai variabel target dari data dan menyimpannya dalam variabel "X"
# Notasi axis=1 digunakan untuk menunjukkan bahwa yang ingin dihapus adalah baris bukan kolom
```

```python
# Membagi data menjadi set pelatihan dan set pengujian
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

#**Logistic Regression**

```python
# Mengimpor pustaka yang diperlukan
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Membangun model regresi logistik
model_lr = LogisticRegression()

# Melatih model pada set pelatihan
model_lr = model_lr.fit(X_train,y_train)
```

```python
# Evaluasi kinerja model data awal
print('Train acccuracy: ', model_lr.score(X_train,y_train))
print('Test accuracy : ', model_lr.score(X_test,y_test))
```

#**Prediction** 

```python
# Melakukan prediksi pada set pengujian
y_pred = model_lr.predict(X_test)
```

```python
pasien_baru = [[2, 3, 2, 1, 65, 2, 2, 2, 2, 2, 2, 2, 1, 1, 2, 1, 3, 2]]
prediksi = model_lr.predict(pasien_baru)

if prediksi == 0:
    print("Pasien baru kemungkinan akan mengalami kematian.")
else:
    print("Pasien baru kemungkinan akan hidup.")
```
